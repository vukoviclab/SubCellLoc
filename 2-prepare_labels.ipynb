```python
import pandas as pd
```

```python
df = pd.read_csv('33_sub_location_labels.csv')
df
```

```python
df[~df.isna().any(axis=1)]
```

```python
df = df[~df['label'].str.contains('-')]
```

```python
df
```

```python
new_df = df.groupby('label')['sub-labels'].agg(list).reset_index()
new_df_transposed = new_df.set_index('label').T
new_df_transposed
```

```python
grouped = df.groupby('label')['sub-labels'].agg(lambda x: list(x)).reset_index()

# Convert the DataFrame to dictionary
data_dict = dict(zip(grouped['label'], grouped['sub-labels']))
data_dict
```

```python
max_len = max(len(lst) for lst in data_dict.values())

# Pad the shorter lists with None
for key in data_dict.keys():
    len_diff = max_len - len(data_dict[key])
    data_dict[key].extend([None]*len_diff)
    
df_final = pd.DataFrame.from_dict(data_dict)
df_final
```

```python
df_final.to_csv('33_sub_location_labels_col.csv', index=False)
```
